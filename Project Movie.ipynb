#Preprocess TITLE_BASICS

remember to use joined data now!

```
# title_basics
title_basics.columns
title_basics.select(title_basics.columns).distinct().display()
```

```
# endYear = 2021
# Ihr könntet die \N Einträge mit der DataFrame.replace() Methode erstmal in tatsächliche None Werte umformen, danach könnt ihr mit col("name").isNull()  relativ einfach filtern (bearbeitet)
# import pyspark.sql.functions as F
#df = df.withColumn(column,F.when(F.isnan(F.col(column)),None).otherwise(F.col(column)))
#title_basics_1 = title_basics.withColumn('endYear',F.when(F.isnan(F.col('endYear')),2021))

# from pyspark.sql.functions import when

#########
# removed import above, because they were already imported
#########

#########
# use joined data set with reduced rows
#########

title_basics_1 =title_basics_ratings.withColumn("endYear", \
              when(title_basics["endYear"] == "\\N", title_basics["startYear"]).otherwise(title_basics["endYear"]))
title_basics_2 =title_basics_1.withColumn("runtimeMinutes", \
              when(title_basics_1["runtimeMinutes"] == "\\N", 1).otherwise(title_basics["runtimeMinutes"]))
title_basics_3 =title_basics_2.withColumn("isAdult", \
              when(title_basics_2["isAdult"] > 1, 1).otherwise(title_basics["isAdult"]))
title_basics_4 = title_basics_3.withColumn("startYear",col("startYear").cast("int")) \
                  .withColumn("endYear",col("endYear").cast("int")) \
                  .withColumn("runtimeMinutes",col("runtimeMinutes").cast("int"))
title_basics_4.dtypes
```

```
title_basics_5 = title_basics_4.join(title_basics_4.select(col("tconst"),split(col("genres"), ",\s*").alias("genres_array")),'tconst')
```

```
# title_basics_5.show()
```

```
# from pyspark.sql.functions import arrays_zip, col, explode

title_basics_6 = \
  (title_basics_5 \
    .withColumn("genres_array", explode("genres_array")))
#    .select("a", col("tmp.b"), col("tmp.c"), "d"))
title_basics_6.show()
```

```
# title_akas_final = title_akas.groupBy(col('titleID')).agg(countDistinct(col('language')).alias("distinct languages"),countDistinct(col('region'))
title_basics_6.groupBy(col("genres_array")).agg(countDistinct(col('tconst')).alias("Count")).display()
```

```
# output columns:
puffer = title_basics_6 \
            .select("genres_array") \
            .distinct() \
            .collect()
```

```
output_cols=[]
for i in range(len(puffer)):
  output_cols.append(puffer[i][0])
output_cols
```

```
### OHE

for name in output_cols:
#name='Animation'
#title_basics_6.select(title_basics_6.tconst, when(title_basics_6.genres_array == name,1).otherwise(0).alias(name)).show()
  title_basics_5 = title_basics_5.select("*", when(title_basics_5.genres.contains(name),1).otherwise(0).alias(name))
  
  
title_basics_5 = title_basics_5.drop('\\N', 'primaryTitle','originalTitle','genres','genres_array').withColumnRenamed('short', 'shortGenre')


# drop runtimeMinutes at first stage
title_basics_5 = title_basics_5.drop('runtimeMinutes')

title_basics_5.printSchema()
#df.select(df.name, F.when(df.age > 3, 1).otherwise(0)).show()
```

```
#df4.groupBy("year").pivot("course", ["dotNET", "Java"]).sum("earnings").collect()

title_basics_7 = title_basics_5 \
                    .join(title_basics_5.groupBy('tconst', 'titleType').pivot('titleType').count().fillna(0),'tconst') \
                    .drop('titleType')
```
