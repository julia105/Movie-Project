```
import numpy as np
import pandas as pd
```

```
def my_info(df):
  info = pd.DataFrame(df.count(), columns=["filled"])
  
 
  info["filled %"] = info["filled"] / len(df)
  info["nunique"] = df.nunique()
  info["dtypes"] = df.dtypes
  info["uniques"] = np.nan
  for idx, row in info.iterrows():
    info.loc[idx, "uniques"] = str(list(df[idx].unique()))
 
  return info
```

```
pd_movie_ratings=movie_ratings.toPandas().drop('tconst', axis=1)
my_info(pd_movie_ratings)
```
